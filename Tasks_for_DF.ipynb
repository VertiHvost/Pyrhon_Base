6. Удаление дубликатов

8. Сортировка
Задача: Отсортируйте DataFrame по Temperature (по убыванию).

```python
df_google = df_google.sort_values('Temperature')
df_google
```

9. Агрегация
Задача: Для каждого города выведите максимум и минимум возраста.

```python
max_light = df_google.groupby("Light")["CO2"].max
min_light = df_google.groupby("Light")["CO2"].min
result = df_google.groupby("Light")["CO2"].agg(['min', 'max']).reset_index()
result
```

10. Визуализация
Задача: Постройте гистограмму для столбца Light.

```python
import matplotlib.pyplot as plt
df_google['Light'].plot.hist(
    bins=10,
    edgecolor='black',
    alpha=0.7,
    figsize=(8,8)
)

plt.title("Гистограмма отношения наблюдения/Light")
plt.xlabel('Light')
plt.ylabel('кол-во наблюдений')
# plt.grid(axis='x', alpha=0.3)

plt.show()
```
